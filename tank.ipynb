```python
class Tank:
    def __init__(self, capacity):
        self.capacity = capacity

        self.current_volume = 0
        self.flow_rate = 0

    def fill(self):
        self.flow_rate = 20

    def drain(self):
        self.flow_rate = -20

    def close_valve(self):
        self.flow_rate = 0

    def update(self):
        if self.current_volume < self.capacity and self.current_volume >= 0:
            self.current_volume += self.flow_rate

    
```

```python
tk1 = Tank(2000)
```

```python
import time

tk1.fill()

try:
    while True:
        tk1.update()
        time.sleep(1)
        print(tk1.current_volume)
except KeyboardInterrupt:
    print("Proceso detenido por el usuario.")
```
